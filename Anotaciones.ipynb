# Instalar o actualizar Python desde los repositorios de Fedora

## sudo dnf install python3

# Crear entornos virtuales con Python

## En Fedora (Linux)

### Crear un entorno virtual


#### python3 -m venv venv

### Activar el entorno

#### source venv/bin/activate


### Desactivar

#### deactivate


## En Windows (CMD o PowerShell)

### Crear el entorno

#### python -m venv mi_entorno


### Activar el entorno

#### CMD

##### mi_entorno\Scripts\activate.bat


#### PowerShell

##### mi_entorno\Scripts\Activate.ps1


##### Si PowerShell bloquea la ejecuci√≥n, activa scripts (una vez):

###### Set-ExecutionPolicy RemoteSigned -Scope CurrentUser

### Desactivar

#### deactivate
